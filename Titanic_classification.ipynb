```python
def family_size(number):
    if number==1:
        return "Alone"
    elif number>1 and number <5:
        return "Small"
    else:
        return "Large"
```

```python
family_size(5)
```

```python
train['family_size']=train['family'].apply(family_size)
```

```python
test['family_size']=test['family'].apply(family_size)
```

```python
train.drop(columns=['family'],inplace=True)
test.drop(columns=['family'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,mean_squared_error
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import roc_curve,roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import CategoricalNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
#testing through decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
yf=clf.predict(Xf)
```

```python
yf.shape
```

```python
dt_cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.set(font_scale=1.2)

sns.heatmap(dt_cm, annot=True, fmt='d', square=True, xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])

plt.xlabel('Predicted Values')
plt.ylabel('True Values')  
plt.title('Confusion Matrix for Decision Tree') 
plt.show()
```

```python
import shap
import matplotlib.pyplot as plt
clf.fit(X_train, y_train)

explainer_of_dtc = shap.TreeExplainer(clf, feature_perturbation="tree_path_dependent")

shap_values = explainer_of_dtc.shap_values(X_test)

shap.summary_plot(shap_values, X_test, plot_type="bar", show=False)

plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

clf.fit(X_train, y_train)

dtc_pred_prob = clf.predict_proba(X_test)[:, 1]

dtc_actual_predict = pd.DataFrame({'y_actual': y_test, 'y_pred_prob': dtc_pred_prob})

fpr, tpr, thresholds = roc_curve(dtc_actual_predict['y_actual'], dtc_actual_predict['y_pred_prob'])

auc = roc_auc_score(dtc_actual_predict['y_actual'], dtc_actual_predict['y_pred_prob'])

plt.plot(fpr, tpr, label=f'Decision Tree (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='grey', label='Random')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Decision Tree')

plt.legend()

plt.show()
```
